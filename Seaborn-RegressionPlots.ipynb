```python
import pandas as pd
import seaborn as sns
flights=sns.load_dataset('flights')
tips=sns.load_dataset('tips')
```

```python
tempData=tips[45:95]
```

```python
tempData.sort_values('total_bill')
```

# lmplot

```python
sns.lmplot(x='total_bill',y='tip',data=tempData)
```

```python
# Categorical addition using hue
```

```python
sns.lmplot(x='total_bill',y='tip',data=tempData,hue='sex')
```

```python
# Markers to have representation with image representable icons like s-square, o- circel, v -triangle
# hue used to represent categories with different color 
# NOTE: ATTENTION TO NEXT EXAMPLE
# scatter_kws to change size of markers i.e. that is functionality of matplot under the seaborn
```

```python
sns.lmplot(x='total_bill',y='tip',data=tempData,hue='sex',markers=['s','v'],scatter_kws={'s':200})
```

```python
# instead of hue try column it will give  separate plots for each category
```

```python
     sns.lmplot(x='total_bill',y='tip',data=tempData,col='day')
```

```python
# Add row as well to the lmplot along with coloumn
```

```python
sns.lmplot(x='total_bill',y='tip',data=tempData,col='day',row='time')
```

```python
# Add row as well to the lmplot along with coloumn, row and hue all together
```

```python
sns.lmplot(x='total_bill',y='tip',data=tempData,col='day',row='time',hue='sex')
```

```python
# Add row as well to the lmplot along with coloumn, row and hue,aspect ratio all together
```

```python
sns.lmplot(x='total_bill',y='tip',data=tempData,col='day',hue='sex',aspect=.8,height=8)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tempData,col='day',hue='sex',aspect=.4,height=8)
```
