```python
import datetime
from pandas_datareader import data,wb
```

```python
start=datetime.datetime(2010,1,1)
end=datetime.datetime(2020,7,23)
```

```python
BAC=data.DataReader('BAC','yahoo',start,end)
Citi=data.DataReader('C','yahoo',start,end)
Google=data.DataReader('GS','yahoo',start,end)
JPM=data.DataReader('JPM','yahoo',start,end)
MS=data.DataReader('MS','yahoo',start,end)
Wells=data.DataReader('WFC','yahoo',start,end)
```

```python
tickers=['BAC','Citi','Google','JPM','MS','Wells']
```

```python
tickers
```

```python
bank_Stocks=pd.concat([BAC,Citi, Google,JPM,MS,Wells],axis=1,keys=tickers)
```

```python
bank_Stocks.columns.names=['Bank','Stock']
```
